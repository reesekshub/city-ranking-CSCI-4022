# City recommender!

This performs a lookup on the association rules to recommend similar cities when provided a list of cities.

```python
input_cities = ["Indianapolis", "Anaheim", "Boulder", "Tucson"] # change this to preferred cities

```

```python
from collections import defaultdict

# Load rules: "ant1, ant2, ... => con1, con2, ..."
# A rule fires when all antecedent cities are in the input list
rules = []
with open("assn_rules.txt") as f:
    for line in f:
        lhs, rhs = line.strip().split("=>")
        antecedents = frozenset(c.strip() for c in lhs.split(","))
        consequents = frozenset(c.strip() for c in rhs.split(","))
        rules.append((antecedents, consequents))

# Count how many rules fire for each candidate city
input_set = set(input_cities)
votes = defaultdict(int)
for antecedents, consequents in rules:
    if antecedents.issubset(input_set):
        for candidate in consequents:
            if candidate not in input_set:
                votes[candidate] += 1

recommendations = sorted(votes.items(), key=lambda x: x[1], reverse=True)

print(f"Recommendations based on: {input_cities}\n")
print(f"{'City':<25} {'Votes':>5}")
print("-" * 32)
for city, count in recommendations[:10]:
    print(f"{city:<25} {count:>5}")
```
